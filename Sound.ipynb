```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

# Phasor

##### if confused, refer to the explanation in Day2 pdf

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 22050       # sampling rate, Hz
dur = 2        # in seconds
freq = 10000.0     # sine frequency, Hz (related to the intensity of sound)
```

```python
# generate time
t = np.arange(1, sr * dur+1)/sr
t
len(t)
```

```python
# generate phase
theta = t * 2*np.pi * freq   
print(theta)
len(theta)
```

```python
# generate signal by cosine-phasor
s = np.sin(theta) # radiant
len(s)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '-.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ipd.Audio(s, rate=sr)
```

##### if confused, refer to the explanation in TransferFunction pdf

```python
# generate signal by complex-phasor
c = np.exp(theta*1j) 
# e^(pi*i)
# exponential

# pi = 0.001... 2pi
# t = 0.001,0.002...1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```
